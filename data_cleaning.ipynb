```python
import plotly.subplots as sp
import plotly.graph_objs as go
import calendar

# Define a dictionary to map day and month names to their corresponding numbers
day_to_num = {day: i for i, day in enumerate(calendar.day_name)}
month_to_num = {month: i for i, month in enumerate(calendar.month_name) if month}

# Add a new column with the corresponding day and month numbers
group_by_market['day_num'] = group_by_market['day'].map(day_to_num)
group_by_market['month_num'] = group_by_market['month'].map(month_to_num)

grouped_by_day = group_by_market.groupby('day').agg({'거래량': 'sum', 'day_num': 'first'}).sort_values(by='day_num')
grouped_by_month = group_by_market.groupby('month').agg({'거래량': 'sum', 'month_num': 'first'}).sort_values(by='month_num')

# Create a subplot layout
fig = sp.make_subplots(rows=1, cols=2, vertical_spacing=0.1, subplot_titles=('거래량_day', '거래량_month'))

# Format numbers with comma separators
day_formatted = [format(int(x), ',') for x in grouped_by_day['거래량']]
month_formatted = [format(int(x), ',') for x in grouped_by_month['거래량']]

# Add histograms for days and months with total numbers displayed on the bars
fig.add_trace(go.Bar(x=grouped_by_day.index, y=grouped_by_day['거래량'], text=day_formatted, textposition='auto', name='Days'), row=1, col=1)
fig.add_trace(go.Bar(x=grouped_by_month.index, y=grouped_by_month['거래량'], text=month_formatted, textposition='auto', name='Months'), row=1, col=2)

# Update layout settings
fig.update_layout(showlegend=False, title_text="거래량 by Day and Month")

# Show the plot
fig.show()
```
